```julia
include("two-level.jl")
using BenchmarkTools
using OrdinaryDiffEq
using PyPlot
```

```julia
function sim_LZ(::Type{T}, npoints=1001; kws...) where T
    sys = T(10, 10, -50, 50, 10)
    ts, ψs = evolve(sys, npoints; kws...)
end
```

```julia
@btime sim_LZ(LabBasisCalc);
@btime sim_LZ(LabBasisCalc, 2);
@btime sim_LZ(LabBasisCalc, alg=Vern7());
@btime sim_LZ(LabBasisCalc, 2, alg=Vern7());
@btime sim_LZ(LabBasisCalc, alg=Vern9());
@btime sim_LZ(LabBasisCalc, 2, alg=Vern9());
```

```julia
r1 = sim_LZ(LabBasisCalc);
r2 = sim_LZ(RotBasisCalc);
qr1 = sim_LZ(QTLabBasisCalc);
qr2 = sim_LZ(QTRotBasisCalc);

figure(figsize=[10, 7.5])
subplot(2, 2, 1)
plot(r1[1], real.(r1[2][1, :]))
plot(r2[1], real.(r2[2][1, :]))
plot(qr1[1], real.(qr1[2][1, :]))
plot(qr2[1], real.(qr2[2][1, :]))
grid()

subplot(2, 2, 2)
plot(r1[1], imag.(r1[2][1, :]))
plot(r2[1], imag.(r2[2][1, :]))
plot(qr1[1], imag.(qr1[2][1, :]))
plot(qr2[1], imag.(qr2[2][1, :]))
grid()

subplot(2, 2, 3)
plot(r1[1], real.(r1[2][2, :]))
plot(r2[1], real.(r2[2][2, :]))
plot(qr1[1], real.(qr1[2][2, :]))
plot(qr2[1], real.(qr2[2][2, :]))
grid()

subplot(2, 2, 4)
plot(r1[1], imag.(r1[2][2, :]))
plot(r2[1], imag.(r2[2][2, :]))
plot(qr1[1], imag.(qr1[2][2, :]))
plot(qr2[1], imag.(qr2[2][2, :]))
grid()
```
