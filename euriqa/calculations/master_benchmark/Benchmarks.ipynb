```julia
using BenchmarkTools
```

```julia
include("qt.jl")
include("qo.jl")
include("simple.jl")
include("opt.jl")
```

```julia
function test_motion(M)
    println(M)
    println("n=10")
    p = M.motion_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 10)
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    p = M.motion2_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 10)
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    println("n=50")
    p = M.motion_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 50)
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    p = M.motion2_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 50)
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    println("n=250")
    p = M.motion_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 250)
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    p = M.motion2_problem(2π * 0.13, 2π * 0.032, 0.0, 0.0, 250)
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2)))
    @btime $(M.solve_motion2)($(p), $(range(0, 4 * π / (2π * 0.13), 2001)))
    return
end
```

```julia
test_motion(QTTest)
```

```julia
test_motion(QOTest)
```

```julia
test_motion(SimpleTest)
```

```julia
test_motion(OPTTest)
```
