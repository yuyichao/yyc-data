```julia
function get_sparse_data(M, H, i0, i1)
    p = M.sparse_problem(H, i0, i1)
    sol = M.psolve(p, range(0, 10, 2001))
    return M.pexpect(p, sol)
end
```

```julia
const P1 = gen_dipole(10, 11, (2.0, 1.2im, -2.1), 0.5, 0.1, 0.05), 11, 33;
const P2 = gen_dipole(10, 9, (3.2im, 2.0, -2.1), -0.5, 0.2, 0.1), 11, 12;
const P3 = gen_dipole(2, 1, (3.2im, -2.1, 0.0), 0.0, 0.1, 0.1), 3, 7;
const P4 = gen_dipole(9, 8, (0.0, 4.0, 0.0), 0.0), 10, 28;
```

```julia
data_qt = get_sparse_data(QTTest, P1...)
data_qo = get_sparse_data(QOTest, P1...)
data_simple = get_sparse_data(SimpleTest, P1...)
data_opt = get_sparse_data(OPTTest, P1...)

figure(figsize=[10, 8])
subplot(2, 2, 1)
plot(data_qt[1], data_qt[2])
plot(data_qo[1], data_qo[2])
plot(data_simple[1], data_simple[2])
plot(data_opt[1], data_opt[2])

subplot(2, 2, 2)
plot(data_qt[1], data_qt[3])
plot(data_qo[1], data_qo[3])
plot(data_simple[1], data_simple[3])
plot(data_opt[1], data_opt[3])

subplot(2, 2, 3)
plot(data_qt[1], data_qt[4])
plot(data_qo[1], data_qo[4])
plot(data_simple[1], data_simple[4])
plot(data_opt[1], data_opt[4])

subplot(2, 2, 4)
plot(data_qt[1], data_qt[5])
plot(data_qo[1], data_qo[5])
plot(data_simple[1], data_simple[5])
plot(data_opt[1], data_opt[5])
```

```julia
function test_sparse(M)
    println(M)
    println("P1")
    p = M.sparse_problem(P1...)
    @btime $(M.psolve)($(p), $(range(0, 10, 2)))
    @btime $(M.psolve)($(p), $(range(0, 10, 2001)))
    println("P2")
    p = M.sparse_problem(P2...)
    @btime $(M.psolve)($(p), $(range(0, 10, 2)))
    @btime $(M.psolve)($(p), $(range(0, 10, 2001)))
    println("P3")
    p = M.sparse_problem(P3...)
    @btime $(M.psolve)($(p), $(range(0, 10, 2)))
    @btime $(M.psolve)($(p), $(range(0, 10, 2001)))
    println("P4")
    p = M.sparse_problem(P4...)
    @btime $(M.psolve)($(p), $(range(0, 10, 2)))
    @btime $(M.psolve)($(p), $(range(0, 10, 2001)))
    return
end
```

```julia
test_sparse(QTTest)
```

```julia
test_sparse(QOTest)
```

```julia
test_sparse(SimpleTest)
```

```julia
test_sparse(OPTTest)
```
